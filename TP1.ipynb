```python
import matplotlib.pyplot as plt;
import numpy as np;
```

An undirected graph is represented by a list whose i-th entry give the neighbors j<i of i

```python
def clique(n):
    return [[j for j in range(i)] for i in range(n)]
```

```python
def randomA(n,p):
    return [[j for j in range(i) if np.random.random()<p] for i in range(n)]
```

```python
randomA(10,0.1)
```

Draw a graph G with the vertices being regularly space on a circle:

```python
def draw(G):
    plt.close('all')
    plt.gca(aspect='equal')
    plt.gca().axis('off')
    n=len(G)
    for k in range(n):
        (xk,yk)=(10*np.cos(2*k*np.pi/n),10*np.sin(2*k*np.pi/n))
        for l in [k]+G[k]: # add k to draw isolated vertices too
                (xl,yl)=(10*np.cos(2*l*np.pi/n),10*np.sin(2*l*np.pi/n))
                plt.plot([xk,xl],[yk,yl],marker = 'o',color='blue')

```

```python
draw(clique(7))
```

```python
draw(randomA(10,0.5))
```

Draws uniformly at random a subset of S of size k:

```python
def randomSubset(S,k):
    if len(S)==0:
        return []
    if np.random.random()<k/len(S):
        return [S[0]]+randomSubset(S[1:],k-1)
    else:
        return randomSubset(S[1:],k)
```

Draws uniformly at random a graph with n vertices and m edges:

```python
def randomB(n,m):
    edges=[]
    for i in range(n):
        for j in range(i):
            edges.append((j,i))
    edges=randomSubset(edges,m)
    return [[j for j in range(i) if (j,i) in edges] for i in range(n)]
```

```python
draw(randomB(20,10))
```

```python
draw(randomB(7, 4))
```
